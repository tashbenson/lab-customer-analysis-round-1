```python
import pandas as pd  
import numpy as np
```

```python
# Read files into python as data frames
# Show Data frame shape
```

```python
file1 = pd.read_csv('file1.csv')
display(file1.head())
print(type(file1))
print(file1.shape)
```

```python
file2 = pd.read_csv('file2.csv')
display(file2.head())
print(type(file2))
print(file2.shape)
```

```python
file3 = pd.read_csv('file3.csv')
display(file3.head())
print(type(file3))
print(file3.shape)
```

```python
# Standardize header names 
col_1 = [] #List to store new column names 
for i in range(len(file1.columns)):
    col_1.append(file1.columns[i].lower()) #Append adds items to the list in lower case
file1.columns = col_1 #Assign new list as new column names
file1.columns = file1.columns.str.replace(" ","_") #Replace blank spaces
file1 = file1.rename(columns={'st':'state'}) #Rename st to state
file1.head() #Check column names
```

```python
col_2 = [] #List to store new column names 
for i in range(len(file2.columns)):
    col_2.append(file2.columns[i].lower()) #Append adds items to the list in lower case
file2.columns = col_2 #Assign new list as new column names
file2.columns = file2.columns.str.replace(" ","_") #Replace blank spaces
file2 = file2.rename(columns={'st':'state'}) #Rename st to state
file2.head() #Check column names
```

```python
col_3 = [] #List to store new column names 
for i in range(len(file3.columns)):
    col_3.append(file3.columns[i].lower()) #Append adds items to the list in lower case
file3.columns = col_3 #Assign new list as new column names
file3.columns = file3.columns.str.replace(" ","_") #Replace blank spaces
file3 = file3.rename(columns={'st':'state'}) #Rename st to state
file3.head() #Check column names
```

```python
#Rearrange columns in dataframes
print(file1.columns)
file1.head()
```

```python
file2=file2[['customer', 'state', 'gender', 'education', 'customer_lifetime_value',
       'income', 'monthly_premium_auto', 'number_of_open_complaints',
       'policy_type', 'vehicle_class', 'total_claim_amount']]
file2.head()
```

```python
file3=file3[['customer', 'state', 'gender', 'education', 'customer_lifetime_value',
       'income', 'monthly_premium_auto', 'number_of_open_complaints',
       'policy_type', 'vehicle_class', 'total_claim_amount']]
file3.head()
```

```python
#Concatenate all three dataframes
print(file1.shape)
print(file2.shape)
print(file3.shape)
```

```python
data = pd.concat([file1, file2, file3], axis=0)
data
```

```python
#Customer - customer number/id
#State
#Gender 
#Education - level of education 
#Customer life time value - estimate of possibility that a customer continues their policy.
#Income 
#Monthly  premium auto - premium the customer pays for insurance
#No. of complaints 
#Policy type - type of insurnace the customer has 
#Vehicle class - type of vehicle
#Total claim amount - amount customer has received for an insurance claim
```

```python
#Delete the column education and the number of open complaints from the dataframe.
data2=data.drop(['education','number_of_open_complaints'], axis=1)
data2.head()
```

```python
#Correct the values in the column customer lifetime value. They are given as a percent, so divide them by 100 and change dtype to numerical type.
data2['customer_lifetime_value'] = data2['customer_lifetime_value'].str.replace('%','')
data2['customer_lifetime_value'] = data2['customer_lifetime_value'].replace(np.nan, 0) #Replace NaN values with 0
data2['customer_lifetime_value']= pd.to_numeric(data2['customer_lifetime_value']) #Change to numerical
data2['customer_lifetime_value'] = data2['customer_lifetime_value']/100
data2.head()
```

```python
#Check for duplicate rows in the data and remove if any.
data2= data2.drop_duplicates()
data2.shape

```

```python
#Filter out the data for customers who have an income of 0 or less.
display(data2[(data2['income']<=100)])
```
